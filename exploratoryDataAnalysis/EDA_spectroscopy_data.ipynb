```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#started the exploratory data analysis
```

```python
df = pd.read_csv('spectro1.csv',sep =",")
df.head(444)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.Label.value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(),cmap='Blues',annot=False)
```

```python
k = 18 #number of variables for heatmap
cols = df.corr().nlargest(k, 'Label')['Label'].index
cm = df[cols].corr()
plt.figure(figsize=(11,8))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```

```python
l = df.columns.values
number_of_columns=18
number_of_rows = len(l)-1/number_of_columns
plt.figure(figsize=(number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.set_style('whitegrid')
    sns.boxplot(df[l[i]],color='green',orient='v')
    plt.tight_layout()
```

```python
plt.figure(figsize=(2*number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.distplot(df[l[i]],kde=True) 
```

```python
new_df=df.groupby(np.arange(len(df))//3).mean()
```

```python
new_df.describe()
```

```python
new_df.to_csv(r'coupled_data.csv')
```

```python
plt.figure(figsize=(11,8))
sns.heatmap(new_df.corr(),cmap='Blues',annot=False)
```

```python
k = 18 #number of variables for heatmap
cols = new_df.corr().nlargest(k, 'Label')['Label'].index
cm = new_df[cols].corr()
plt.figure(figsize=(11,8))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```

```python
l = new_df.columns.values
number_of_columns=18
number_of_rows = len(l)-1/number_of_columns
plt.figure(figsize=(number_of_columns,5*number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.set_style('whitegrid')
    sns.boxplot(new_df[l[i]],color='green',orient='v')
    plt.tight_layout()
```
